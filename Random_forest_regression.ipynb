# Random Forest Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestRegressor #model
```

## Importing the dataset

```
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values

dataset
```

## Training the Random Forest Regression model on the whole dataset

```
regressor = RandomForestRegressor(n_estimators=10,random_state=0)  #n estimatores é essencial

regressor.fit(X,y)
```

## Predicting a new result

```
regressor.predict([[6.5]]) #NICE!
```

## Visualising the Random Forest Regression results (higher resolution)

```
#2 steps from position level(red points). using only Decision Tree was 1 step... More trees...

X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))

plt.scatter(X,y,color='red')
plt.plot(X_grid,regressor.predict(X_grid),color='blue')  #transformed matrix of features x powers
plt.title('Truth of Bluff (Random Forest Regression Tree)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show
```
